```python
class PriorityQueueADT:
    def __init__( self ):
        self.data = []
    
    def is_empty( self ):
        return len( self.data ) == 0
    
    def length( self ):
        return len( self.data )
    
    def enqueue( self, prioridad, value ):
        dato = [prioridad, value]
        contador = -1
        continuar = True

        if self.is_empty(): #Si la lista está vacía, agregamos el primer elemento 
            self.data.append( dato )

        else: 
            while continuar:
                contador += 1
                
                if contador == self.length(): #Si hemos llehado al último elemento
                    continuar = False

                elif dato[0] < self.data[contador][0]: #Si nos encontramos un numero de prioridad más alto, detenemos
                    continuar = False

            self.data.insert(contador, dato)

    def dequeue( self ):
        return self.data.pop(0)
    
    def to_string( self ):
        for elem in self.data:
            print(f"|{elem[0]}, {elem[1]}| <-- ", end= "")
        print("")
```

```python
print("--- PRUEBA DE PriorityQueueADT ---")
cola = PriorityQueueADT()

print("\n¿Está vacío? R=", cola.is_empty())
print("Numero de elementos:", cola.length())

print("\nEjercicio: Evacue el barco utilizando la Cola con Prioridad No Acotada: ")

cola.enqueue(4, "Mecanico")
cola.to_string()
cola.enqueue(2, "Niños")
cola.to_string()
cola.enqueue(5, "Capitan")
cola.to_string()
cola.enqueue(4, "Timonel")
cola.to_string()
cola.enqueue(4, "Vigia")
cola.to_string()
cola.enqueue(4, "Maestre")
cola.to_string()
cola.enqueue(2, "3ra Edad")
cola.to_string()
cola.enqueue(3, "Mujeres")
cola.to_string()
cola.enqueue(1, "Niñas")
cola.to_string()
cola.enqueue(3, "Hombres")
cola.to_string()

print("\nLista de prioridad de evacuación final: ")
cola.to_string()
print("¿Está vacío? R=", cola.is_empty())
print("Numero de elementos:", cola.length())

siguiente = cola.dequeue()
print("\nProximo en ser evacuado: ", siguiente)
siguiente = cola.dequeue()
print("Proximo en ser evacuado: ", siguiente)
siguiente = cola.dequeue()
print("Proximo en ser evacuado: ", siguiente)
siguiente = cola.dequeue()
print("Proximo en ser evacuado: ", siguiente)
siguiente = cola.dequeue()
print("Proximo en ser evacuado: ", siguiente)

print("\nPersonas restantes en espera: ")
cola.to_string()
```

```python
class QueueADT:
    def __init__( self ):
        self.data = []
    
    def is_empty( self ):
        return len( self.data ) == 0
    
    def length( self ):
        return len( self.data )
    
    def enqueue( self, value ):
        self.data.append( value )
    
    def dequeue( self ):
        return self.data.pop(0)
    
    def to_string( self ):
        for elem in self.data:
            print(f"|{elem}| <-- ", end= "")
        print("")
```

```python
class BoundedPriorityQueueADT: 
    def __init__( self, niveles ):
        self.data = [QueueADT() for x in range(niveles) ]
        self.size = 0
        self.niveles = niveles

    def is_empty( self ):
        if self.size == 0:
            return True
        else:
            return False

    def length( self ):
        return self.size

    def enqueue( self, prioridad, value ):
        if prioridad < 0 or prioridad > self.niveles:
            print(f"Prioridad NO válida. Ingrese una prioridad entre 1 y {self.niveles}")
        else:
            self.data[prioridad - 1].enqueue( value )
            self.size += 1

    def dequeue( self ):
        if self.size != 0:

            i = 0
            while self.data[i].is_empty():  
                i += 1
            
            self.size -= 1
            return self.data[i].dequeue()
            
        else:
            print("Cola vacía, no es posible sacar el primer elemento")

    
    def to_string( self ):
        for i in range(self.niveles):
            print(f"Prioridad {i + 1}: ",end = "")
            self.data[i].to_string()
```

```python
print("--- PRUEBA DE BoundedPriorityQueueADT ---")
acotada = BoundedPriorityQueueADT( 5 )
print("\n¿Está vacía? R=", acotada.is_empty() )
print("Numero de elementos:", acotada.length())

print("")
acotada.enqueue(4, "Mecanico")
acotada.enqueue(2, "Niños")
acotada.enqueue(5, "Capitan")
acotada.enqueue(4, "Timonel")
acotada.enqueue(4, "Vigia")
acotada.enqueue(4, "Maestre")
acotada.enqueue(2, "3ra Edad")
acotada.enqueue(3, "Mujeres")
acotada.enqueue(1, "Niñas")
acotada.enqueue(3, "Hombres")
acotada.to_string()
print("\n¿Está vacía? R=", acotada.is_empty() )
print("Numero de elementos:", acotada.length())

print("\nIniciando la evuación...")
print("Evacuando:",acotada.dequeue())
print("Evacuando:",acotada.dequeue())
print("Evacuando:",acotada.dequeue())
print("Evacuando:",acotada.dequeue())
print("Evacuando:",acotada.dequeue())
print("\nPersonas esperando evacuación: ")
acotada.to_string()
print("\n¿Está vacía? R=", acotada.is_empty() )
print("Numero de elementos:", acotada.length())
```
